## 1. Гармонический сигнал

<img src="./src/1.png">

### Сигнал

```julia
s(t, A, f, ϕ) = A*sin(2pi*f*t+ϕ);
```

### Параметры

```julia
A, ϕ = 1, 0.5
dt = 0.5e-6
time = collect(0:dt:0.1)
freqs = 1e3*collect(1:10:100)
tn = length(time)
fn = length(freqs);
```

### Расчет

```julia
res = Array{Float64}(undef,tn,2,fn)
for (i,vT) in enumerate(time), (j,vF) in enumerate(freqs)
    res[i, 1, j] = vT
    res[i, 2, j] = s(vT, A, vF, ϕ)
end
```

### Визуализация

```julia
lim = 1e-3
animH = @animate for i in 1:fn
    tempfreq = Int(1e-3*freqs[i])
    plt = plot(res[:,1,i],res[:,2,i], 
    xlims=(0,lim), xticks = 0:lim/2:lim,
    title="Гармонический сигнал", lab = "f = $tempfreq кГц",
    xlabel="Время, с", ylabel="Уровень, произв. ед.")
    savefig(plt,"./graphs/harmonic/harmonicFREQ=$tempfreq.png")
end
gif(animH, "./graphs/harmonic/mainHarmonic.gif", fps = 0.5)
```

## 2. Периодический сигнал

<img src="./src/2.png">

### Сигнал

```julia
sQ(t, A, P, dC) = ((mod(t,P)/P)>dC) ? A : 0;
```

### Параметры и расчет

```julia
A, tp, P = 1, 0.5e-3, 1e-3
dC = tp/P
dt = 1e-5
time = collect(0:dt:1e-2)
parms = time, A, P, dC
tn = length(time)
sig = sQ.(parms...);
```

### Получение спектра

```julia
Fs = 1/dt
freqsF = 1e-3*rfftfreq(tn, Fs) # kHz
ampF = rfft(sig); 
```

### График

```julia
tdPlot = plot(time,sig,
leg=false,title="Последовательность прямоугольных импульсов", xlabel="Время, с", ylabel="Уровень, произв. ед.")
fdPlot = plot(freqsF,abs.(ampF),
leg=false,title="Амплитудный спектр", xlabel="Частота, кГц", ylabel="Уровень, произв. ед.")
figP = plot(tdPlot,fdPlot, layout=(2,1), plot_title="Периодический сигнал")
savefig(figP,"./graphs/periodic/main.png")
display(figP)
```

## 3. Амплитудная модуляция

<img src="./src/3.png">

### Сигнал и модуляция

```julia
car(t,A,f) = A*sin(2pi*f*t) 
msg(t,A,f,ϕ) = A*cos(2pi*f*t+ϕ) 
am(t,m,Acar,Amsg,fcar,fmsg,ϕmsg) = car(t,Acar,fcar)*(1+m*msg(t,Amsg,fmsg,ϕmsg)/(Amsg)); 
```

### Функции для расчета и визуализации

Получение спектра и исходные графики

```julia
function getSpec(t,s,dt)
    tn = length(t)
    fs = 1/dt
    freqs = 1e-3*rfftfreq(tn, fs) # kHz
    amps = abs.(rfft(s)) # dB
    return freqs, amps
end;
```

```julia
function figInitial(t,dt,m,Acar,Amsg,fcar,fmsg,ϕmsg)
    sigC = car.(t,Acar,fcar)
    sigM = msg.(t,Amsg,fmsg,ϕmsg)
    sigAM = am.(t,m,Acar,Amsg,fcar,fmsg,ϕmsg)
    freqs, amps = getSpec(t,sigAM,dt)
    limX = 1e-3
    pltC = plot(t,sigC,xlims=(0,limX), xticks = 0:limX/2:limX, yticks=(-Amsg:Amsg),
        leg=false, title="Несущая", xlabel="Время, с", ylabel="Уровень, произв. ед.")
    pltM = plot(t,sigM,xlims=(0,limX), xticks = 0:limX/2:limX, yticks=(-Amsg:Amsg),
        leg=false, title="Информационный сигнал", xlabel="Время, с", ylabel="Уровень, произв. ед.")
    pltAM = plot(t,sigAM,xlims=(0,limX), xticks = 0:limX/2:limX, yticks=(-Amsg:Amsg),
        leg=false, title="Модулированный сигнал", xlabel="Время, с", ylabel="Уровень, произв. ед.")
    limY = 1000
    pltSpec = plot(freqs,abs.(amps),xlims=(0,200), ylims=(0,limY), yticks=(0:limY/2:limY),
        leg=false, title="Спектр модулированного сигнала", xlabel="Частота, кГц", ylabel="Уровень,  произв. ед.")
    fig = plot(pltC,pltM,pltAM,pltSpec,layout=(4,1), plot_title="Амплитудная модуляция",
        ytickfontsize=5, yguidefontsize=5)
    savefig(fig,"./graphs/am/initial.png")
    return fig
end;
```

Функции для варьирования параметров и получения графиков

```julia
function figVarCarF(t,dt,m,Acar,Amsg,carFV,fmsg,ϕmsg)
    tn, pn, iterT, iterP = length(t), length(carFV), enumerate(t), enumerate(carFV)
    specn = div(tn,2)+1
    sigArray = Array{Float64}(undef,tn,pn)
    specArray = Array{Float64}(undef,specn,2,pn)
    sigPlotArray = Vector{Plots.Plot}(undef,pn)   
    specPlotArray = Vector{Plots.Plot}(undef,pn)
    for (i,vT) in iterT, (j,vP) in iterP
        label = 1e-3*vP
        sigArray[i,j]=am(vT,m,Acar,Amsg,vP,fmsg,ϕmsg)
        limX = 1e-3
        sigPlotArray[j] = plot(t,sigArray[:,j],xlims=(0,limX), xticks = 0:limX/2:limX, 
        lab="fc = $label кГц", title="Модулированный сигнал", xlabel="Время, с", ylabel="Уровень, произв. ед.")
    end
    for i in 1:pn
        specArray[:,1,i], specArray[:,2,i] = getSpec(t,sigArray[:,i],dt)
        limX, limY = 250, 1000
        specPlotArray[i] = plot(specArray[:,1,i], specArray[:,2,i], leg=false, 
        xlims=(0,limX), xticks = 0:limX/2:limX, ylims=(0,limY), yticks = 0:limY/2:limY, 
        title = "Амплитудный спектр", xlabel="Частота, кГц", ylabel="Уровень,  произв. ед.")
    end
    animAMCarF = @animate for i in 1:pn
        name = carFV[i]
        plt = plot(sigPlotArray[i],specPlotArray[i],layout=(2,1), plot_title="Варьирование частоты несущей при АМ")
        savefig(plt,"./graphs/am/carf/carf=$name.png")
    end
    pltgif = gif(animAMCarF, "./graphs/am/carf/mainCarF.gif", fps = 0.5)
    return pltgif
end

function figVarMsgF(t,dt,m,Acar,Amsg,fcar,msgFV,ϕmsg)
    tn, pn, iterT, iterP = length(t), length(msgFV), enumerate(t), enumerate(msgFV)
    specn = div(tn,2)+1
    sigArray = Array{Float64}(undef,tn,pn)
    specArray = Array{Float64}(undef,specn,2,pn)
    sigPlotArray = Vector{Plots.Plot}(undef,pn)   
    specPlotArray = Vector{Plots.Plot}(undef,pn)
    for (i,vT) in iterT, (j,vP) in iterP
        label = 1e-3*vP
        sigArray[i,j]=am(vT,m,Acar,Amsg,fcar,vP,ϕmsg)
        limX = 1e-3
        sigPlotArray[j] = plot(t,sigArray[:,j],xlims=(0,limX), xticks = 0:limX/2:limX, 
        lab="fm = $label кГц", title="Модулированный сигнал", xlabel="Время, с", ylabel="Уровень, произв. ед.")
    end
    for i in 1:pn
        specArray[:,1,i], specArray[:,2,i] = getSpec(t,sigArray[:,i],dt)
        limX, limY = 250, 1000
        specPlotArray[i] = plot(specArray[:,1,i], specArray[:,2,i], leg=false, 
        xlims=(0,limX), xticks = 0:limX/2:limX, ylims=(0,limY), yticks = 0:limY/2:limY, 
        title = "Амплитудный спектр", xlabel="Частота, кГц", ylabel="Уровень,  произв. ед.")
    end
    animAMMsgF = @animate for i in 1:pn
        name = msgFV[i]
        plt = plot(sigPlotArray[i],specPlotArray[i],layout=(2,1), plot_title="Варьирование частоты модуляции при АМ")
        savefig(plt,"./graphs/am/msgf/msgf=$name.png")
    end
    pltgif = gif(animAMMsgF, "./graphs/am/msgf/mainMsgF.gif", fps = 0.5)
    return pltgif
end

function figVarM(t,dt,mV,Acar,Amsg,fcar,fmsg,ϕmsg)
    tn, pn, iterT, iterP = length(t), length(mV), enumerate(t), enumerate(mV)
    specn = div(tn,2)+1
    sigArray = Array{Float64}(undef,tn,pn)
    specArray = Array{Float64}(undef,specn,2,pn)
    sigPlotArray = Vector{Plots.Plot}(undef,pn)   
    specPlotArray = Vector{Plots.Plot}(undef,pn)
    for (i,vT) in iterT, (j,vP) in iterP
        label = vP
        sigArray[i,j]=am(vT,vP,Acar,Amsg,fcar,fmsg,ϕmsg)
        limX = 1e-3
        sigPlotArray[j] = plot(t,sigArray[:,j],xlims=(0,limX), xticks = 0:limX/2:limX, 
        lab="m = $label", title="Модулированный сигнал", xlabel="Время, с", ylabel="Уровень, произв. ед.")
    end
    for i in 1:pn
        specArray[:,1,i], specArray[:,2,i] = getSpec(t,sigArray[:,i],dt)
        limX, limY = 250, 1000
        specPlotArray[i] = plot(specArray[:,1,i], specArray[:,2,i], leg=false, 
        xlims=(0,limX), xticks = 0:limX/2:limX, ylims=(0,limY), yticks = 0:limY/2:limY, 
        title = "Амплитудный спектр", xlabel="Частота, кГц", ylabel="Уровень,  произв. ед.")
    end
    animAMM = @animate for i in 1:pn
        name = mV[i]
        plt = plot(sigPlotArray[i],specPlotArray[i],layout=(2,1), plot_title="Варьирование глубины модуляции при АМ")
        savefig(plt,"./graphs/am/m/m=$name.png")
    end
    pltgif = gif(animAMM, "./graphs/am/m/mainM.gif", fps = 0.5)
    return pltgif
end;
```

## 4. Частотная модуляция

<img src="./src/4.png">

### Сигнал и модуляция

```julia
wv(t, A, f) = A*sin(2pi*f*t)
fm(t, ACar, AMsg, fCar, fMsg, fDev) = ACar*cos((2pi*fCar*t)+fDev/fMsg*AMsg*sin(2pi*fMsg*t));
```

### Функции для расчета и визуализации

```julia
function figFMInitial(t,dt,ACar,AMsg,fCar,fMsg,fDev)
    car = wv.(t, ACar, fCar)
    msg = wv.(t, AMod, fMsg)
    sig = fm.(t, ACar, AMsg, fCar, fMsg, fDev)
    freqs, amps = getSpec(t,sig,dt)
    limX, limY = 1e-3, AMsg
    pCar = plot(t,car,leg=false, xlims=(0,limX), xticks = 0:limX/2:limX, yticks=(-limY:limY),
    title="Несущая", xlabel="Время, с", ylabel="Уровень, произв. ед.") 
    pMsg = plot(t,msg,leg=false, xlims=(0,limX), xticks = 0:limX/2:limX, yticks=(-limY:limY),
    title="Информационный сигнал", xlabel="Время, с", ylabel="Уровень, произв. ед.")
    pSig = plot(t,sig,leg=false, xlims=(0,limX), xticks = 0:limX/2:limX, yticks=(-limY:limY),
    title="Модулированный сигнал", xlabel="Время, с", ylabel="Уровень, произв. ед.")
    limX, limY = 200, 500
    specPlot = plot(freqs, amps, leg=false, xlims=(0,limX), xticks = 0:limX/2:limX, ylims=(0,limY), yticks = 0:limY/2:limY,
    title = "Амплитудный спектр", xlabel="Частота, кГц", ylabel="Уровень,  произв. ед.")
    fig = plot(pCar, pMsg, pSig, specPlot, layout=(4,1),
    plot_title="Частотная модуляция", ytickfontsize=5, guidefontsize=5)
    savefig(fig,"./graphs/fm/initial.png")
    return fig
end

function figVarMsgF(t,dt,Acar,Amsg,fcar,msgFV,fDev)
    tn, pn, iterT, iterP = length(t), length(msgFV), enumerate(t), enumerate(msgFV)
    specn = div(tn,2)+1
    sigArray = Array{Float64}(undef,tn,pn)
    specArray = Array{Float64}(undef,specn,2,pn)
    sigPlotArray = Vector{Plots.Plot}(undef,pn)   
    specPlotArray = Vector{Plots.Plot}(undef,pn)
    for (i,vT) in iterT, (j,vP) in iterP
        label = 1e-3*vP
        sigArray[i,j]=fm(vT,Acar,Amsg,fcar,vP,fDev)
        limX = 1e-3
        sigPlotArray[j] = plot(t,sigArray[:,j],xlims=(0,limX), xticks = 0:limX/2:limX, 
        lab="fm = $label кГц", title="Модулированный сигнал", xlabel="Время, с", ylabel="Уровень, произв. ед.")
    end
    for i in 1:pn
        specArray[:,1,i], specArray[:,2,i] = getSpec(t,sigArray[:,i],dt)
        limX, limY = 250, 500
        specPlotArray[i] = plot(specArray[:,1,i], specArray[:,2,i], leg=false, 
        xlims=(0,limX), xticks = 0:limX/2:limX, ylims=(0,limY), yticks = 0:limY/2:limY, 
        title = "Амплитудный спектр", xlabel="Частота, кГц", ylabel="Уровень,  произв. ед.")
    end
    animAMMsgF = @animate for i in 1:pn
        name = msgFV[i]
        plt = plot(sigPlotArray[i],specPlotArray[i],layout=(2,1), plot_title="Варьирование частоты модуляции при ЧМ")
        savefig(plt,"./graphs/fm/msgf/fmsg=$name.png")
    end
    pltgif = gif(animAMMsgF, "./graphs/fm/msgf/mainMsgF.gif", fps = 0.5)
    return pltgif
end;
```

### Параметры

Константы

```julia
ACar, AMod = 1, 1
fCar, fMsgInitial = 1e3.*(100, 10)
fDev = 1e3*50
dt = 0.5e-6
time = collect(0:dt:1e-3);
```

Список для варьирования частоты модуляции

```julia
msgFV = 1e3*collect(range(1,25,3));
```

Наборы параметров

```julia
pInitial = time, dt, ACar, AMod, fCar, fMsgInitial, fDev
pVarMsgF = time, dt, ACar, AMod, fCar, msgFV, fDev;
```

### Расчет и визуализация

```julia
figVarMsgF(pVarMsgF...)
```

## 5. Фазовая и частотная модуляция

<img src="./src/5.png">

### Фазовая модуляция

```julia
pm(t, Acar, Amod, fcar, fmod, k) = Amod*cos(2pi*fmod*t+k*wv(t, Acar, fcar));
```

### Функция расчета и получения графиков  

```julia
function figFMPM(t,dt,Acar,Amsg,fcar,msgFV,fDev,kPM)
    gr(size=(1600,1200))
    tn, pn, iterT, iterP = length(t), length(msgFV), enumerate(t), enumerate(msgFV)
    specn = div(tn,2)+1
    car = wv.(t,Acar,fcar)
    limCarX, limCarY = 1e-6, Acar
    carPlot = plot(t,car,leg=false, xlims=(0,limCarX), xticks = 0:limCarX/2:limCarX, ylims=(-limCarY,limCarY), yticks = -limCarY:limCarY/2:limCarY,
    title="Несущая", xlabel="Время, с")
    modArray, modPlotArray = Array{Float64}(undef,tn,pn), Vector{Plots.Plot}(undef,pn)
    sigArray, sigPlotArray = Array{Float64}(undef,tn,pn,2), Array{Plots.Plot}(undef,pn,2)
    specArray, specPlotArray = Array{Float64}(undef,specn,2,pn,2), Array{Plots.Plot}(undef,pn,2)
    for (i,vT) in iterT, (j,vP) in iterP
        modArray[i,j] = wv(vT,Amsg,vP) 
        sigArray[i,j,1] = fm(vT,Acar,Amsg,fcar,vP,fDev)
        sigArray[i,j,2] = pm(vT,Acar,Amsg,fcar,vP,kPM)
    end

    for i in 1:pn
        label = 1e-3*msgFV[i]
        limModX, limModY = 1e-4, Amsg
        modPlotArray[i] = plot(t,modArray[:,i],lab="fm=$label кГц", 
        xlims=(0,limModX), xticks = 0:limModX/2:limModX, ylims=(-limModY,limModY), yticks = -limModY:limModY/2:limModY, 
        title="Информационный сигнал", xlabel="Время, с")
        limSigX, limSigY = 1e-4, Acar 
        sigPlotArray[i,1] = plot(t,sigArray[:,i,1],leg=false,
        xlims=(0,limSigX), xticks = 0:limSigX/2:limSigX, ylims=(-limSigY,limSigY), yticks = -limSigY:limSigY/2:limSigY,
        title="Частотная модуляция", xlabel="Время, с")
        sigPlotArray[i,2] = plot(t,sigArray[:,i,2],leg=false,
        xlims=(0,limSigX), xticks = 0:limSigX/2:limSigX, ylims=(-limSigY,limSigY), yticks = -limSigY:limSigY/2:limSigY,
        title="Фазовая модуляция", xlabel="Время, с")
        freqsFM, ampsFM = getSpec(t,sigArray[:,i,1],dt)
        freqsPM, ampsPM = getSpec(t,sigArray[:,i,2],dt)
        specArray[:,1,i,1], specArray[:,2,i,1] = 1e-3.*freqsFM, ampsFM # MHz
        specArray[:,1,i,2], specArray[:,2,i,2] = 1e-3.*freqsPM, ampsPM # MHz
        #limSpecX = 5e2
        #xlims=(0,limSpecX), xticks = 0:limSpecX/2:limSpecX,
        specPlotArray[i,1] = plot(specArray[:,1,i,1], specArray[:,2,i,1],leg=false,        
        xlabel="Частота, МГц")
        specPlotArray[i,2] = plot(specArray[:,1,i,2], specArray[:,2,i,2],leg=false,
        xlabel="Частота, МГц")
    end  
    animFMPM = @animate for i in 1:pn
        name = msgFV[i]
        car = carPlot
        mod = modPlotArray[i]
        sigFM, sigPM = sigPlotArray[i,1], sigPlotArray[i,2]
        specFM, specPM = specPlotArray[i,1], specPlotArray[i,2]
        l = @layout[a;b;grid(2,2)]
        pAll = plot(car,mod,sigFM,sigPM,specFM,specPM,layout=l, 
        plot_title="Частотная и фазовая модуляция",ytickfontsize=5, guidefontsize=5)
        savefig(pAll,"./graphs/fmpm/fmsg=$name.png")
    end
    fmpmGIF = gif(animFMPM, "./graphs/fmpm/main.gif", fps = 0.5)
    return fmpmGIF 
end;
```

###  Параметры

Константы

```julia
ACar, AMsg = 1, 1
fCar = 1e6*100
fDev = 1e3*50
kPM = 5
dt = 1e-9
time = collect(0:dt:1e-4);
```

Список для варьирования частоты модуляции

```julia
msgVF = 1e3.*(1, 5, 10);
```

Список параметров

```julia
pFMPM = time,dt,ACar,AMsg,fCar,msgVF,fDev,kPM;
```

### Расчет и визуализация
